```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert
```

```python
fs = 1000
N = 1000
t = np.arange(N) / fs

f = [10, 30]
ff = np.linspace(f[0], np.mean(f), N)

chirp = np.sin(2 * np.pi * ff * t)

fCoeffs = np.fft.rfft(chirp)
hz = np.fft.rfftfreq(N, 1 / fs)

z = hilbert(chirp)
phase = np.unwrap(np.angle(z))
inst_freq = np.diff(phase) * fs / (2*np.pi)
inst_t = (t[:-1] + t[1:]) / 2

plt.figure(figsize=(21, 16))
plt.subplot(331)
plt.plot(t, chirp)
plt.ylabel("Amp")
plt.xlabel("time")
plt.title("Time")

plt.subplot(332)
plt.plot(np.abs(fCoeffs))
plt.ylabel("Amp")
plt.xlabel("hz")
plt.title("Freq")
plt.xlim(-0.1, 96)

plt.subplot(333)
plt.plot(z.real, z.imag)
plt.xlabel("Real")
plt.ylabel("Imag")
plt.title("Hilbert real-imag")

plt.subplot(334)
plt.plot(t, np.abs(z))
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("Hilbert abs")

plt.subplot(335)
plt.plot(t, np.angle(z))
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("Hilbert Angle")

plt.subplot(336)
plt.plot(t, phase)
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("unwrap Hilbert Angle")

plt.subplot(337)
plt.plot(t[1:], inst_freq)
plt.xlabel("Time")
plt.ylabel("HZ")
plt.title("Inst Freq.")

plt.tight_layout()
```
