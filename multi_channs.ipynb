```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 500
N = 100
C = 10
n = np.arange(N)
t = n / fs
hz = np.fft.fftfreq(N, 1/fs)

data = np.zeros((C, N))

for i in range(C):
    print(i * 5)
    data[i] = np.sin(2 * np.pi * i * 5 * t) + np.sin(2 * np.pi * 10 * t)
    
plt.figure(figsize=(15, 4))

plt.subplot(121)
for i in range(C):
    plt.plot(data[i] + (i * 4))
plt.title("Good stuff")
    
# calculate FFT
fCoeffs = np.abs(np.fft.fft(data, axis=1))

plt.subplot(122)
plt.stem(hz, fCoeffs[9])
plt.xlim(-1, 50)
```

```python
plt.imshow(fCoeffs.real)
```
