```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
srate = 500
t = np.linspace(0, 1, srate)

amp1 = 1
amp2 = 1

f1 = 5
f2 = 5

theta1 = np.pi/2 * 5
theta2 = np.pi / 2 * 0

signal1 = amp1 * np.sin(2 * np.pi * f1 * t + theta1)
signal2 = amp2 * np.sin(2 * np.pi * f2 * t + theta2)

print(np.dot(signal1, signal2))

plt.plot(t, signal1, 'r')
plt.plot(t, signal2, 'b--')
```

```python
srate = 400
t = np.linspace(-2, 2, srate)
f = 3
theta = np.pi / 2
pure = np.sin(2 * np.pi * f * t + theta)
noise = np.exp(-(t**2) / 0.1)

morlet = pure * noise

plt.figure(figsize=(11, 7))

plt.subplot(231)
plt.plot(t, pure)
plt.title("Pure Signal")

plt.subplot(232)
plt.plot(t, noise)
plt.title("Noise Signal")

plt.subplot(233)
plt.plot(t, morlet)
plt.title("Morlet")

freqs = np.arange(0, 20)
spectral_analysis_real = []
spectral_analysis_imag = []

for freq in freqs:
    comparative_sig_real = np.cos(2 * np.pi * freq * t)
    comparative_sig_imag = np.sin(2 * np.pi * freq * t)

    dotprod_real = np.abs(np.dot(comparative_sig_real, morlet))
    dotprod_imag = np.abs(np.dot(comparative_sig_imag, morlet))

    spectral_analysis_real.append(dotprod_real)
    spectral_analysis_imag.append(dotprod_imag)

plt.subplot(234)
plt.stem(freqs, spectral_analysis_real)
plt.title("Spectral Analysis Real")

plt.subplot(235)
plt.stem(freqs, spectral_analysis_imag)
plt.title("Spectral Analysis Imag")

plt.subplot(236)
plt.stem(freqs, np.abs(np.array(spectral_analysis_imag) + np.array(spectral_analysis_real)))
plt.title("Spectral Analysis Imag")

plt.tight_layout()
```
