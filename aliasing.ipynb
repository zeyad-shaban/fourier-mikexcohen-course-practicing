```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fm = 10
def fun(t):
    return np.sin(2 * np.pi * fm * t)


N = 100
# nyquist freq at fs/2, our fm is 5, fm >= fs/2 otherwise it si impossible to sample.. .well that looks like a lie lol
fs = fm * 2
continuous_t = np.linspace(0, 10, 5000)
t = np.arange(0 + 1/fm / 4, 10, 1/fs)

continuous_wave = fun(continuous_t)
wave = fun(t)

plt.plot(continuous_t, continuous_wave, label='haha')
plt.plot(t, wave, label='haha')
plt.plot(t, wave, 'ro', label='haha')
plt.xlim(0, 1)
```
